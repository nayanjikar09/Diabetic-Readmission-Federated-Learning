```python
import matplotlib.pyplot as plt

# Data
models = ['Hospital A', 'Hospital B', 'Hospital C', 'Global Federated']
accuracy = [74.40, 74.88, 74.75, 79.07]

# Create bar chart
plt.figure(figsize=(8, 5))
plt.bar(models, accuracy)

# Labels and title
plt.xlabel('Model')
plt.ylabel('Accuracy (%)')
plt.title('Figure 10: Accuracy Comparison Plot')

# Display accuracy values on top of bars
for i, value in enumerate(accuracy):
    plt.text(i, value + 0.2, f'{value}%', ha='center')

# Set y-axis limit for better visualization
plt.ylim(70, 82)

# Show plot
plt.tight_layout()
plt.show()
```
