```python
import pandas as pd

files = {
    "Hospital A": r"E:\semm 8\federated_learning\federated_3\a-synthetic data\A_synthetic_diabetic_data.csv",
    "Hospital B": r"E:\semm 8\federated_learning\federated_3\b-synthetic data\B_synthetic_diabetic_data.csv",
    "Hospital C": r"E:\semm 8\federated_learning\federated_3\c-synthetic data\C_synthetic_diabetic_data_v3.csv"
}

rows = []

for name, path in files.items():
    df = pd.read_csv(path)
    dist = df["readmitted"].value_counts(normalize=True) * 100
    rows.append([
        name,
        round(dist.get("NO", 0), 2),
        round(dist.get(">30", 0), 2),
        round(dist.get("<30", 0), 2)
    ])

table = pd.DataFrame(
    rows,
    columns=["Client", "NO (%)", ">30 (%)", "<30 (%)"]
)

print(table)
```
