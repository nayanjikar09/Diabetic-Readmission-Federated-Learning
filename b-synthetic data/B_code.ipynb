```python
import pandas as pd
import numpy as np

# Load original data
df = pd.read_csv(r"E:\semm 8\federated_learning\federated_3\data_preprosses\preprocessed_diabetic_data.csv")

# Number of synthetic samples
n_samples = len(df)

# Bootstrap resampling (with replacement)
synthetic_df = df.sample(
    n=n_samples,
    replace=True,
    random_state=92
).reset_index(drop=True)

# Save synthetic data
synthetic_df.to_csv("B_synthetic_diabetic_data.csv", index=False)

print("100% distribution-matching synthetic data generated!")

```
